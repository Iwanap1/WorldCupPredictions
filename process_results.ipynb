```python
import json
import pandas as pd
```

```python
with open("predictions.json", "r") as f:
    predictions = json.load(f)
```

```python
fixtures = pd.read_csv("data/2026/fixtures_preprocessed.csv")
```

```python
fixtures.head()
```

```python
results = []
for fid in predictions["IwanModel"].keys():
    fid = int(fid)
    fixture = fixtures[fixtures["id"] == fid]
    results.append({
        "team1": fixture["team_1_name"].values[0],
        "team2": fixture["team_2_name"].values[0],
        "IwanModel_t1_score": predictions["IwanModel"][str(fid)]["team_1_score"],
        "IwanModel_t2_score": predictions["IwanModel"][str(fid)]["team_2_score"],
        "AngusModel_t1_score": predictions["AngusModel"][str(fid)]["team_1_score"],
        "AngusModel_t2_score": predictions["AngusModel"][str(fid)]["team_2_score"]
    })
```

```python
res = pd.DataFrame(results)
res.to_csv("results.csv")
```
