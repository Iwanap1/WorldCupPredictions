```python
import pandas as pd
```

```python
fixtures_df = pd.read_csv("data/2026/matches.csv")
teams_df = pd.read_csv("data/2026/teams.csv")
```

```python
fixtures_df.head()
```

```python
teams_df.head()
```

```python
for index, team_row in teams_df.iterrows():
    team_id = team_row["id"]
    name = team_row["team_name"]
    fixtures_df.loc[fixtures_df["home_team_id"] == team_id, "team_1_name"] = name
    fixtures_df.loc[fixtures_df["away_team_id"] == team_id, "team_2_name"] = name
```

```python
fixtures_df.dropna().to_csv("data/2026/fixtures_preprocessed.csv", index=False)
```

```python
fixture_df_teams = fixtures_df.dropna()["team_1_name"].unique().tolist()
fixture_df_teams
```

```python
import json

with open("iwan/preprocessed_data/2026_featurised_squads.json", "r") as f:
    featurised_squads = json.load(f)
```

```python
expected_teams = list(featurised_squads.keys())
for t in fixture_df_teams:
    if t not in expected_teams:
        print(t)
```
